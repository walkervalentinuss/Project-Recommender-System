## Data Pre-processong

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data/Tempat-Wisata-Toba-Preprocessing.csv")
```

```python
data = data.rename(columns={'ReviewerId': 'user_id', 'PlaceID': 'item_id', 'Rating': 'rating'})
# data['user_id'] = data['user_id'].astype(str)
# data['item_id'] = data['item_id'].astype(str)
data['rating'] = data['rating'].astype(float)
data.dropna(inplace=True)
data.head(5)
```

```python
# Step 1: Split into 85% training+validation and 15% test
train_val_data, test_data = train_test_split(data, test_size=0.15, random_state=42)
# Step 2: Define validation size as 17% of train+validation, rounded up
val_size = int(len(train_val_data) * 0.17) + 1
# Step 3: Split 85% data into training (about 70% of original) and validation (about 15% of original)
train_data, val_data = train_test_split(train_val_data, test_size=val_size, random_state=42)
# Check sizes
print(f"Train data size: {train_data.shape}")
print(f"Validation data size: {val_data.shape}")
print(f"Test data size: {test_data.shape}")
```

   - train_val_data sebagai data validasi.

3. Membagi Training+Validation Menjadi Training dan Validation:

## Model Pembanding CF - SVD

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
def create_user_item_matrix(data, user_col='user_id', item_col='item_id', rating_col='rating'):
    data = data.groupby([user_col, item_col])[rating_col].mean().reset_index()
    return data.pivot(index=user_col, columns=item_col, values=rating_col).fillna(0)

# Apply the function to create matrices for each split
train_matrix = create_user_item_matrix(train_data)
val_matrix = create_user_item_matrix(val_data).reindex(columns=train_matrix.columns, fill_value=0)
test_matrix = create_user_item_matrix(test_data).reindex(columns=train_matrix.columns, fill_value=0)
```

Penjelasan Fungsi create_user_item_matrix
1. Parameter Fungsi:
   - data: DataFrame yang berisi informasi user_id, item_id, dan rating.
   - user_col: Nama kolom untuk ID pengguna, defaultnya 'user_id'.
   - item_col: Nama kolom untuk ID item, defaultnya 'item_id'.
   - rating_col: Nama kolom untuk rating, defaultnya 'rating'.

2. Langkah di dalam Fungsi:
   - data.groupby([user_col, item_col])[rating_col].mean().reset_index():
     - Langkah ini mengelompokkan data berdasarkan user_id dan item_id, lalu menghitung rata-rata rating dari setiap pasangan pengguna-item.
     - Jika ada beberapa rating untuk pasangan pengguna-item tertentu, maka nilai rating rata-rata akan digunakan.
   - data.pivot(index=user_col, columns=item_col, values=rating_col).fillna(0):
     - Fungsi pivot digunakan untuk mengonversi data menjadi format user-item matrix, dengan user_id sebagai indeks baris dan item_id sebagai kolom.
     - fillna(0): Mengisi nilai yang hilang dalam matriks dengan 0. Hal ini berarti bahwa jika pengguna belum memberikan rating untuk suatu item, maka nilai defaultnya adalah 0.

3. Mengaplikasikan Fungsi ke Set Data Training, Validation, dan Test:
   - train_matrix: Matriks user-item untuk data training.
   - val_matrix: Matriks user-item untuk data validation. Matriks ini disesuaikan agar kolomnya sama dengan train_matrix menggunakan reindex, dengan nilai default 0 untuk item yang mungkin tidak ada di data validasi tetapi ada di data training.
   - test_matrix: Matriks user-item untuk data test. Juga disesuaikan agar kolomnya sama dengan train_matrix dengan reindex.rix dengan reindex.

Analisis Kode
1. Manfaat Representasi User-Item Matrix:
   - Matriks ini memudahkan model rekomendasi, terutama model berbasis dekomposisi matriks (seperti SVD) untuk menemukan pola tersembunyi dalam data rating.
   - Dengan menyimpan data dalam format matriks, kita dapat mengaplikasikan algoritma dekomposisi matriks untuk menemukan faktor-faktor laten (latent factors) yang mendasari hubungan antara pengguna dan item.

2. Mengisi Nilai Kosong dengan 0:
   - Menggunakan fillna(0) memiliki makna bahwa pengguna yang belum menilai item tertentu dianggap memberikan rating 0. Ini bisa relevan dalam kasus tertentu, tetapi ada pendekatan lain yang mengisi nilai kosong dengan nilai rata-rata atau nilai lain yang lebih bermakna.
   - Pengisian dengan 0 juga membantu algoritma untuk berfokus pada rating yang nyata tanpa terpengaruh oleh nilai yang hilang.

3. Penggunaan reindex:
reindex memastikan bahwa val_matrix dan test_matrix memiliki struktur kolom yang sama dengan train_matrix. Ini penting agar model rekomendasi dapat bekerja dengan matriks yang konsisten dan memastikan bahwa semua item yang muncul di data training juga terwakili di data validation dan test, meskipun pengguna di validation dan test mungkin belum menilai item tersebut.m tersebut.

```python
# Creating an SVD model
def svd_model(train_matrix, n_components=50):
    svd = TruncatedSVD(n_components=n_components, random_state=42)
    # Latih model SVD pada matriks latih
    svd.fit(train_matrix)
    return svd
```

```python
# Train the SVD model
svd_model_trained = svd_model(train_matrix)
```

Analisis Kode
1. Manfaat Penggunaan SVD dalam Sistem Rekomendasi:
   - SVD membantu mendekomposisi matriks user-item yang besar dan jarang (sparse) menjadi ruang laten dengan dimensi yang lebih kecil, di mana setiap komponen laten menangkap hubungan tersembunyi antara pengguna dan item.
   - Model rekomendasi yang dihasilkan dapat memberikan saran yang lebih akurat, karena faktor laten ini membantu mengidentifikasi pola preferensi pengguna terhadap fitur-fitur umum dari item yang direkomendasikan.

2. Pemilihan n_components:
   - Pemilihan nilai n_components merupakan aspek penting dalam model SVD. Jumlah komponen laten yang optimal bergantung pada kompleksitas data dan kebutuhan model.
   - Nilai yang lebih rendah mungkin tidak menangkap cukup informasi, sementara nilai yang terlalu tinggi dapat menyebabkan model menjadi kompleks dan overfitting.
   - Dalam kasus ini, 50 komponen digunakan sebagai titik awal, dan nilai ini dapat disesuaikan atau diuji lebih lanjut untuk menemukan yang paling optimal berdasarkan performa model di data validasi.

3. Dekomposisi Matriks dan Ruang Laten:
   - Proses ini mengubah data dari matriks besar yang mungkin memiliki banyak nilai kosong menjadi representasi yang lebih padat dalam ruang laten, yang dapat digunakan untuk menemukan kemiripan antara pengguna dan item.
   - Setelah dekomposisi, model dapat merekomendasikan item dengan cara mengidentifikasi faktor-faktor laten yang memiliki keterkaitan dengan pengguna tertentu.

### Matrix Evaluation

```python
# Project Test and Validation Matrix
def project_matrix(matrix, svd_model):
    matrix_reduced = svd_model.transform(matrix)
    
    matrix_approx = svd_model.inverse_transform(matrix_reduced)
    return matrix_approx

val_matrix_approx = project_matrix(val_matrix, svd_model_trained)
test_matrix_approx = project_matrix(test_matrix, svd_model_trained)
```

1. Proyeksi Matriks dengan SVD:
   - Tujuan dari proyeksi ini adalah untuk memperkirakan rating pada matriks validasi dan test menggunakan representasi laten yang ditemukan dari data training.
   - Dengan memproyeksikan ke ruang laten, model dapat menghasilkan perkiraan rating berdasarkan pola umum yang ditemukan dalam data latih.

2. Penggunaan transform dan inverse_transform:
   - transform mengonversi data ke dalam bentuk laten yang lebih kompak dan kaya informasi, memungkinkan kita untuk menangkap hubungan antar pengguna dan item dalam dimensi yang lebih rendah.
   - inverse_transform mengembalikan representasi ini ke bentuk aslinya dalam user-item matrix, tetapi dengan nilai yang diprediksi berdasarkan representasi laten tersebut.

3. Kegunaan val_matrix_approx dan test_matrix_approx:
   - Matriks val_matrix_approx dan test_matrix_approx yang dihasilkan akan memiliki perkiraan rating yang dapat digunakan untuk mengevaluasi model pada data validasi dan test.ang sesuai.

```python
# Function to calculate RMSE
def compute_rmse(true_matrix, approx_matrix):
    return np.sqrt(mean_squared_error(true_matrix.ravel(), approx_matrix.ravel()))

# Functions for RMSE and MAE evaluation
def evaluate_rmse_mae(true_matrix, approx_matrix):
    # Convert to numpy array for evaluation
    true_matrix = true_matrix.values if isinstance(true_matrix, pd.DataFrame) else true_matrix
    approx_matrix = approx_matrix.values if isinstance(approx_matrix, pd.DataFrame) else approx_matrix
    rmse = compute_rmse(true_matrix, approx_matrix)
    mae = mean_absolute_error(true_matrix.ravel(), approx_matrix.ravel())
    return rmse, mae

# Function to project a matrix and return a DataFrame
def project_matrix_to_df(matrix, svd_model, columns):
    matrix_df = pd.DataFrame(matrix, columns=columns)
    matrix_reduced = svd_model.transform(matrix_df)
    matrix_approx = svd_model.inverse_transform(matrix_reduced)
    return pd.DataFrame(matrix_approx, columns=columns)
```

```python
val_matrix_approx_df = project_matrix_to_df(val_matrix, svd_model_trained, val_matrix.columns)
test_matrix_approx_df = project_matrix_to_df(test_matrix, svd_model_trained, test_matrix.columns)

# Evaluation on validation and test matrices
val_rmse, val_mae = evaluate_rmse_mae(val_matrix, val_matrix_approx_df)
test_rmse, test_mae = evaluate_rmse_mae(test_matrix, test_matrix_approx_df)
```

```python
# Output of evaluation results
print(f"Validation RMSE: {val_rmse}")
print(f"Validation MAE: {val_mae}")
print(f"Test RMSE: {test_rmse}")
print(f"Test MAE: {test_mae}")
```

Analisis Hasil
1. RMSE dan MAE yang Rendah:
   - RMSE maupun MAE pada data validasi dan test memiliki nilai yang sangat rendah. Ini menunjukkan bahwa prediksi model cukup akurat dalam mendekati nilai rating asli.
   - RMSE rendah menunjukkan bahwa rata-rata deviasi kuadrat dari prediksi terhadap rating asli kecil, sehingga model mampu meminimalkan kesalahan besar.
   - MAE rendah menunjukkan bahwa rata-rata kesalahan absolut (perbedaan langsung antara prediksi dan nilai asli) juga kecil, yang berarti bahwa model ini sangat baik dalam memprediksi nilai rating yang hampir tepat.

2. Konsistensi antara Validation dan Test:
   - Nilai RMSE dan MAE pada data validasi dan test sangat mirip, menunjukkan bahwa model memiliki generalization error yang rendah. Artinya, performa model pada data yang belum terlihat (data test) tidak jauh berbeda dengan performanya pada data validasi.
   - Ini menunjukkan bahwa model tidak mengalami overfitting pada data latih, sehingga model yang dihasilkan mampu bekerja dengan baik pada data baru.

```python
def mean_average_precision_at_k(true_matrix, approx_matrix, k=5):
    map_score = 0
    total_users = true_matrix.shape[0]
    
    for user in range(total_users):
        true_ratings = true_matrix.iloc[user].values
        approx_ratings = approx_matrix.iloc[user].values
        
        # Get top-k recommendations based on predictions
        top_k_items = np.argsort(approx_ratings)[::-1][:k]
        relevant_items = np.argsort(true_ratings)[::-1][:k]  # item yang relevan berdasarkan rating asli
        
        # Calculate precision at each level k
        hits = len(set(top_k_items) & set(relevant_items))
        if hits > 0:
            map_score += hits / k
    
    return map_score / total_users if total_users > 0 else 0
```

```python
# Calculate MAP on validation and test data
val_map = mean_average_precision_at_k(val_matrix, val_matrix_approx_df, k=5)
test_map = mean_average_precision_at_k(test_matrix, test_matrix_approx_df, k=5)

print(f"Validation MAP: {val_map}")
print(f"Test MAP: {test_map}")
```

Analisis Hasil
1. Interpretasi Nilai MAP:
   - MAP pada data validasi dan test berada di sekitar 0.226. Ini menunjukkan bahwa, rata-rata, sekitar 22.6% dari rekomendasi top-5 yang diberikan oleh model adalah relevan (sesuai dengan item yang mendapat rating tinggi pada data asli).
   - Nilai MAP yang mendekati 1 menunjukkan performa yang sangat baik, sedangkan nilai yang lebih rendah menunjukkan bahwa model kurang akurat dalam memberikan rekomendasi item relevan di posisi teratas.

2. Konsistensi antara Validation dan Test:
   - Nilai MAP pada data validasi dan test cukup konsisten. Ini menunjukkan bahwa model tidak overfitting pada data latih dan mampu menjaga performanya pada data yang belum terlihat (data test).
   - Konsistensi ini juga menunjukkan bahwa model dapat memberikan rekomendasi yang cukup relevan pada data baru dengan tingkat presisi yang mirip seperti pada data validasi.


```python
def precision_at_k(true_matrix, approx_matrix, k=5, threshold=4.0):
    precision_scores = []

    for user in range(true_matrix.shape[0]):
        # Original ratings and predictions for a particular user
        true_ratings = true_matrix.iloc[user].values
        approx_ratings = approx_matrix.iloc[user].values

        # Get top-k items with the highest predicted rating
        top_k_items = np.argsort(approx_ratings)[::-1][:k]
        
        # Count the number of relevant items in top-k predictions
        relevant_items = [i for i in top_k_items if true_ratings[i] >= threshold]
        precision = len(relevant_items) / k
        precision_scores.append(precision)
    
    return np.mean(precision_scores)
```

```python
def recall_at_k(true_matrix, approx_matrix, k=5, threshold=4.0):
    recall_scores = []

    for user in range(true_matrix.shape[0]):
        # Original ratings and predictions for a particular user
        true_ratings = true_matrix.iloc[user].values
        approx_ratings = approx_matrix.iloc[user].values

        # Get the actual relevant items
        relevant_items = [i for i in range(len(true_ratings)) if true_ratings[i] >= threshold]
        if len(relevant_items) == 0:
            continue

        # Get top-k items with the highest predicted rating
        top_k_items = np.argsort(approx_ratings)[::-1][:k]
        
        # Count the number of relevant items in top-k predictions
        hits = len(set(top_k_items) & set(relevant_items))
        recall = hits / len(relevant_items)
        recall_scores.append(recall)
    
    return np.mean(recall_scores)
```

```python
# Calculate Precision and Recall on validation data and test data
val_precision = precision_at_k(val_matrix, val_matrix_approx_df, k=5, threshold=4.0)
test_precision = precision_at_k(test_matrix, test_matrix_approx_df, k=5, threshold=4.0)

val_recall = recall_at_k(val_matrix, val_matrix_approx_df, k=5, threshold=4.0)
test_recall = recall_at_k(test_matrix, test_matrix_approx_df, k=5, threshold=4.0)
```

```python
print(f"Validation Precision: {val_precision}")
print(f"Test Precision: {test_precision}")
print(f"Validation Recall: {val_recall}")
print(f"Test Recall: {test_recall}")
```

Analisis Hasil
1. Precision Rendah pada Top-5:
   - Nilai Precision pada data validasi dan test berada di sekitar 0.196, yang berarti bahwa sekitar 19.6% dari rekomendasi top-5 yang diberikan oleh model adalah item yang relevan (item yang sesuai dengan rating >= 4).
   - Precision rendah menunjukkan bahwa dari 5 rekomendasi teratas, hanya sedikit yang benar-benar sesuai dengan item yang disukai oleh pengguna. Ini menunjukkan bahwa model perlu ditingkatkan dalam merekomendasikan item yang relevan pada posisi teratas.

2. Recall Tinggi pada Top-5:
   - Nilai Recall sangat tinggi, yaitu sekitar 0.94 pada data validasi dan 0.947 pada data test.
   - Recall tinggi menunjukkan bahwa model berhasil menemukan hampir semua item relevan di antara semua item yang dianggap relevan oleh pengguna dalam top-5 rekomendasi.
   - Hal ini menunjukkan bahwa meskipun Precision rendah, model cukup baik dalam menemukan item relevan untuk pengguna, meskipun beberapa di antaranya mungkin tidak berada pada posisi teratas.

Precision yang rendah dan Recall yang tinggi menunjukkan bahwa model memiliki coverage yang baik (Recall tinggi) tetapi kurang akurat dalam menempatkan item relevan di peringkat teratas (Precision rendah).
Ini bisa menjadi indikasi bahwa model cenderung memberikan lebih banyak rekomendasi yang mungkin tidak relevan di posisi teratas, meskipun ia cukup baik dalam mengenali item yang disukai oleh pengguna secara keseluruhan.

```python
def get_recommendations_for_user(user_id, approx_matrix, original_matrix, n_recommendations=5):# Pastikan user_id adalah string jika diperlukan
    user_id = str(user_id)

    # Make sure the matrix index is a string
    approx_matrix.index = approx_matrix.index.astype(str)
    original_matrix.index = original_matrix.index.astype(str)

    # Check if user_id is in the original matrix index
    if user_id not in original_matrix.index:
        print(f"User ID {user_id} tidak ditemukan dalam data.")
        return None

    # Get rating prediction for a given user
    user_pred_ratings = approx_matrix.loc[user_id].values
    user_true_ratings = original_matrix.loc[user_id].values

    # Create a DataFrame from the original prediction and rating results.
    recommendations_df = pd.DataFrame({
        'item_id': original_matrix.columns,
        'predicted_rating': user_pred_ratings,
        'actual_rating': user_true_ratings
    })

    # Filter items that users haven't rated yet (recommend only new items)
    recommendations_df = recommendations_df[recommendations_df['actual_rating'] == 0]

    # Sort items by highest predicted rating
    recommendations_df = recommendations_df.sort_values(by='predicted_rating', ascending=False)

    # Take top-n recommendations
    top_recommendations = recommendations_df.head(n_recommendations)

    return top_recommendations

recommendations = get_recommendations_for_user(user_id, test_matrix_approx_df, test_matrix, n_recommendations=5)

# Displays recommendation results
if recommendations is not None:
    print(f"Top 5 rekomendasi untuk user {user_id}:")
    print(recommendations)
```
