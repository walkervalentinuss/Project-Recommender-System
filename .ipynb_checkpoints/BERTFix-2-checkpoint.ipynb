## Data Pre-processing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Tempat-Wisata-Toba-Preprocessing.csv")
```

```python
data.shape
```

```python
data.columns
```

```python
data = data.rename(columns={'ReviewerId': 'user_id', 'PlaceID': 'item_id', 'Rating': 'rating'})
data['user_id'] = data['user_id'].astype(str)
data['item_id'] = data['item_id'].astype(str)
data['rating'] = data['rating'].astype(float)
data.dropna(inplace=True)
data.head(5)
```

```python
data.shape
```

```python
# Step 1: Split into 85% training+validation and 15% test
train_val_data, test_data = train_test_split(data, test_size=0.15, random_state=42)
# Step 2: Define validation size as 17% of train+validation, rounded up
val_size = int(len(train_val_data) * 0.17) + 1
# Step 3: Split 85% data into training (about 70% of original) and validation (about 15% of original)
train_data, val_data = train_test_split(train_val_data, test_size=val_size, random_state=42)
# Check sizes
print(f"Train data size: {train_data.shape}")
print(f"Validation data size: {val_data.shape}")
print(f"Test data size: {test_data.shape}")
```

Penjelasan Langkah-Langkah

1. Membagi Data Menjadi Training+Validation dan Test:
   - train_test_split(data, test_size=0.15, random_state=42): Fungsi ini digunakan untuk membagi dataset awal menjadi dua bagian, yaitu:
     - 85% untuk Training+Validation (train_val_data)
     - 15% untuk Test (test_data)
   - Pembagian ini bertujuan untuk menyisihkan sebagian data sebagai data uji akhir (test data), yang akan digunakan untuk mengevaluasi kinerja model setelah tahap pelatihan dan tuning selesai.

2. Menentukan Ukuran Data Validation:
   - val_size = int(len(train_val_data) * 0.17) + 1:
     - Setelah membagi data awal menjadi training+validation dan test, kita menentukan bahwa 17% dari train_val_data akan digunakan sebagai data validasi.
     - Penambahan +1 memastikan bahwa ukuran validasi selalu dibulatkan ke atas, agar kita mendapatkan sekitar 17% dari
   - train_val_data sebagai data validasi.

3. Membagi Training+Validation Menjadi Training dan Validation:
   - train_test_split(train_val_data, test_size=val_size, random_state=42): Fungsi ini membagi train_val_data menjadi dua bagian:
     - Training Data (train_data), sekitar 70% dari data asli.
     - Validation Data (val_data), sekitar 15% dari data asli.
   - Pembagian ini bertujuan untuk menyediakan data validasi yang akan digunakan untuk tuning model, misalnya dalam memilih hyperparameter.

4. Menampilkan Ukuran Setiap Subset:
   - print(f"Train data size: {train_data.shape}"): Menampilkan ukuran data training.
   - print(f"Validation data size: {val_data.shape}"): Menampilkan ukuran data validasi.
   - print(f"Test data size: {test_data.shape}"): Menampilkan ukuran data uji.ilkan ukuran data uji.

Analisis Pembagian Data
1. Pembagian data yang dilakukan di sini cukup standar dalam machine learning dan sangat penting untuk memastikan bahwa model tidak overfitting atau underfitting.
2. Training Set (70%) digunakan untuk melatih model, mempelajari pola-pola dalam data, dan mengoptimalkan parameter model.
3. Validation Set (15%) menyediakan data yang terpisah dari training untuk tuning hyperparameter dan memilih model terbaik. Ini mencegah model dari overfitting pada training data karena kita melakukan evaluasi selama proses tuning pada data yang belum pernah dilihat oleh model.
4. Test Set (15%) adalah bagian dari data yang sepenuhnya terpisah, digunakan untuk mengevaluasi performa akhir model setelah proses pelatihan dan tuning selesai. Ini memberikan gambaran yang lebih objektif tentang bagaimana model akan berkinerja pada data baru atau data di dunia nyata.

## Model Pembanding CBF - (BERT)

```python
from transformers import BertTokenizer, BertModel  # Mengimpor BertTokenizer dan BertModel dari transformers
import torch  # Mengimpor library PyTorch untuk tensor dan percepatan komputasi dengan GPU
from sklearn.metrics.pairwise import cosine_similarity  # Mengimpor fungsi cosine_similarity untuk menghitung kesamaan antar-vektor
import pandas as pd  # Mengimpor pandas untuk manipulasi dan analisis data terstruktur
import numpy as np  # Mengimpor numpy untuk operasi numerik dan array
```

```python
# Step 1: Load and preprocess dataset
import pandas as pd  # Mengimpor pandas untuk memanipulasi dan menganalisis data terstruktur

# Path to the dataset
file_path_new = 'Tempat-Wisata-Toba-Preprocessing.csv'  # Menentukan lokasi file dataset

# Load the dataset
dataset_new = pd.read_csv(file_path_new)  # Membaca dataset dari file CSV ke dalam DataFrame pandas

# Display the first few rows to verify
dataset_new.head()  # Menampilkan beberapa baris pertama dari dataset untuk memverifikasi keberhasilan pembacaan data
```

```python
# Menggabungkan kolom 'Nama_tempat_wisata' dan 'Reviews' untuk membentuk deskripsi konten
# Jika kolom 'Reviews' memiliki nilai NaN (kosong), maka akan diisi dengan string  ('belum diulas')
dataset_new['description'] = dataset_new['Nama_tempat_wisata'] + " " + dataset_new['Reviews'].fillna('belum diulas')  

dataset_new.head()  # Verifikasi hasil penggabungan kolom dalam dataset
```

```python
# Drop duplicates based on 'PlaceID'
dataset_new = dataset_new.drop_duplicates(subset='PlaceID')

# Display the dataset to verify changes
dataset_new.head()
```

```python
# Step 2: Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')  # Memuat tokenizer BERT pre-trained untuk teks lowercase
bert_model = BertModel.from_pretrained('bert-base-uncased')  # Memuat model BERT pre-trained untuk menghasilkan embedding teks
```

```python
# Step 3: Function to get BERT embeddings for text
def get_bert_embedding(text):  # Mendefinisikan fungsi untuk mendapatkan embedding BERT dari teks
    # Tokenisasi teks dengan tokenizer BERT
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True, padding="max_length")  # Mengembalikan tensor PyTorch
    with torch.no_grad():  # Tidak menghitung gradien untuk mempercepat proses (mode evaluasi)
        outputs = bert_model(**inputs)  # Mendapatkan output dari model BERT
    embedding = outputs.last_hidden_state[:, 0, :].squeeze(0)  # Mengambil embedding dari token CLS (index pertama)
    return embedding.numpy()  # Mengembalikan embedding sebagai array NumPy

# Step 4: Compute embeddings for each place
dataset_new['embedding'] = dataset_new['description'].apply(get_bert_embedding)  # Menghitung embedding untuk setiap deskripsi dan menyimpannya dalam kolom baru 'embedding'

# Display the updated dataset
dataset_new.head()  # Menampilkan beberapa baris pertama dari dataset yang telah diperbarui
```

```python
# Step 4: Compute embeddings for each place
dataset_new['embedding'] = dataset_new['description'].apply(get_bert_embedding)  # Menghitung embedding BERT untuk setiap deskripsi tempat wisata dan menyimpannya di kolom 'embedding'

# Display the updated dataset
dataset_new.head()  # Menampilkan beberapa baris pertama dari dataset yang telah diperbarui
```

```python
# Step 5: Build similarity matrix
embeddings = np.array(dataset_new['embedding'].tolist())  # Mengubah kolom 'embedding' menjadi array NumPy untuk semua deskripsi
similarity_matrix = cosine_similarity(embeddings)  # Menghitung matriks kesamaan menggunakan cosine similarity antar embedding

# Verify the similarity matrix
similarity_matrix.shape  # Mengecek ukuran matriks kesamaan untuk memastikan hasilnya benar
```

```python
# Step 6: Recommendation function
def recommend_places(place_id, dataset, similarity_matrix, top_n=5):  # Fungsi untuk merekomendasikan tempat wisata
    try:
        # Mendapatkan indeks tempat wisata berdasarkan PlaceID
        place_index = dataset[dataset['PlaceID'] == place_id].index[0]
        
        # Mengambil skor kesamaan untuk tempat wisata tersebut
        similarity_scores = similarity_matrix[place_index]
        
        # Mengurutkan skor kesamaan dan mengambil indeks tempat dengan skor tertinggi (kecuali tempat itu sendiri)
        top_indices = similarity_scores.argsort()[-top_n-1:-1][::-1]  # Mengambil top_n skor tertinggi
        
        # Mendapatkan rekomendasi berdasarkan indeks yang diperoleh
        recommendations = dataset.iloc[top_indices]
        
        # Mengembalikan kolom tertentu untuk rekomendasi
        return recommendations[['PlaceID', 'Nama_tempat_wisata', 'Category']]
    except IndexError:  # Jika PlaceID tidak ditemukan dalam dataset
        return "Place ID not found in the dataset."
```

```python
# Ensure dataset_new and similarity_matrix are defined
place_id = 0  # Tentukan PlaceID yang valid dari dataset
recommendations = recommend_places(place_id, dataset_new, similarity_matrix, top_n=5)  # Memanggil fungsi rekomendasi dengan PlaceID tertentu

if isinstance(recommendations, str):  # Memeriksa apakah hasil rekomendasi adalah string (error message)
    print(recommendations)  # Jika ada pesan error, cetak pesan tersebut
else:
    print(f"Top recommendations for PlaceID {place_id}:\n", recommendations)  # Jika berhasil, cetak daftar rekomendasi
```

```python
# Menampilkan grafik untuk rekomendasi (perbaikan)
def plot_similarity(place_id, recommendations, similarity_matrix, dataset):
    try:
        # Mendapatkan indeks PlaceID yang diberikan
        place_index = dataset[dataset['PlaceID'] == place_id].index[0]
        
        # Mengambil skor kesamaan tempat rekomendasi berdasarkan indeks relatif
        recommendation_indices = recommendations.index.to_list()  # Mendapatkan indeks rekomendasi
        relative_indices = [dataset.index.get_loc(idx) for idx in recommendation_indices]  # Menyesuaikan dengan indeks matriks kesamaan
        
        scores = similarity_matrix[place_index, relative_indices]  # Mengambil skor kesamaan dari matriks
        
        # Nama tempat wisata
        recommended_names = recommendations['Nama_tempat_wisata'].values
        
        # Plot
        plt.figure(figsize=(10, 6))
        plt.barh(recommended_names, scores, color='skyblue')  # Bar horizontal
        plt.xlabel("Similarity Score")  # Label sumbu X
        plt.title(f"Top Similarity Scores for PlaceID {place_id}")  # Judul grafik
        plt.gca().invert_yaxis()  # Membalik urutan bar agar yang teratas sesuai
        plt.show()
    except IndexError as e:
        print(f"Error: {e}. Please check PlaceID or dataset indexing.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Jika ada rekomendasi, tampilkan grafik
if not isinstance(recommendations, str):  # Pastikan rekomendasi berhasil ditemukan
    plot_similarity(place_id, recommendations, similarity_matrix, dataset_new)
else:
    print(recommendations)  # Tampilkan pesan error jika rekomendasi tidak ditemukan
```
